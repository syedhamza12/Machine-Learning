```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
sf=load_iris()
dir(sf)
```

```python
#here we make irisi data frame
dfm=pd.DataFrame(sf.data,columns=sf.feature_names)
dfm.head()
```

```python
dfm['taget']=sf.target
dfm.head()
```

```python
dfm['flower']=dfm.taget.apply(lambda x :sf.target_names[x] )
dfm.head()
```

```python
x_h=dfm.drop(['taget','flower'],axis='columns')
x_h.head()
```

```python
y_h=dfm.taget
y_h.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x_h,y_h,test_size=0.2)
```

```python
from sklearn.svm import SVC
mods=SVC()
```

```python
mods.fit(x_h,y_h)
```

```python
mods.predict(X_test)
```
